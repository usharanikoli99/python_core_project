```python
# Python Bank System with Loan Feature

accounts = {}
next_account_number = 1001

# Create a new account
def create_account():
    global next_account_number
    name = input("Enter your name: ")
    pin = input("Set a 4-digit PIN: ")
    deposit = float(input("Enter initial deposit: ₹"))

    accounts[str(next_account_number)] = {
        "name": name,
        "pin": pin,
        "balance": deposit
    }

    print(f"\nAccount created successfully!")
    print(f"Account Number: {next_account_number}")
    print("PIN set. Please keep it safe.\n")

    next_account_number += 1

# Login to account
def login():
    acc_no = input("Enter your account number: ")
    pin = input("Enter your 4-digit PIN: ")

    if acc_no in accounts and accounts[acc_no]["pin"] == pin:
        print(f"\nWelcome, {accounts[acc_no]['name']}!\n")
        user_menu(acc_no)
    else:
        print("Invalid account number or PIN.\n")

# Deposit money
def deposit(acc_no):
    amount = float(input("Enter deposit amount: ₹"))
    accounts[acc_no]["balance"] += amount
    print(f"₹{amount} deposited. New Balance: ₹{accounts[acc_no]['balance']}\n")

# Withdraw money
def withdraw(acc_no):
    amount = float(input("Enter withdrawal amount: ₹"))
    if amount <= accounts[acc_no]["balance"]:
        accounts[acc_no]["balance"] -= amount
        print(f"₹{amount} withdrawn. Remaining Balance: ₹{accounts[acc_no]['balance']}\n")
    else:
        print("Insufficient balance.\n")

# Check balance
def check_balance(acc_no):
    print(f"Current Balance: ₹{accounts[acc_no]['balance']}\n")

# Change PIN
def change_pin(acc_no):
    old_pin = input("Enter your current PIN: ")
    if old_pin == accounts[acc_no]["pin"]:
        new_pin = input("Enter new 4-digit PIN: ")
        accounts[acc_no]["pin"] = new_pin
        print("PIN changed successfully!\n")
    else:
        print("Incorrect PIN.\n")

# Delete account
def delete_account(acc_no):
    confirm = input("Are you sure you want to delete your account? (yes/no): ")
    if confirm.lower() == "yes":
        del accounts[acc_no]
        print("Account deleted successfully.\n")
        return True
    return False

# Apply for loan
def apply_loan(acc_no):
    print("Loan Application Form:")
    amount = float(input("Enter loan amount you need: ₹"))
    reason = input("Reason for loan: ")

    print("\nLoan Approved!")
    accounts[acc_no]["balance"] += amount
    print(f"₹{amount} has been added to your account.")
    print("Loan Summary:")
    print(f"Loan Amount: ₹{amount}")
    print(f"Loan Reason: {reason}")
    print(f"New Balance: ₹{accounts[acc_no]['balance']}\n")

# Menu after login
def user_menu(acc_no):
    while True:
        print("1. Deposit")
        print("2. Withdraw")
        print("3. Check Balance")
        print("4. Change PIN")
        print("5. Delete Account")
        print("6. Apply for Loan")
        print("7. Logout")

        choice = input("Choose an option: ")
        print()

        if choice == '1':
            deposit(acc_no)
        elif choice == '2':
            withdraw(acc_no)
        elif choice == '3':
            check_balance(acc_no)
        elif choice == '4':
            change_pin(acc_no)
        elif choice == '5':
            if delete_account(acc_no):
                break
        elif choice == '6':
            apply_loan(acc_no)
        elif choice == '7':
            print("Logged out.\n")
            break
        else:
            print("Invalid choice.\n")

# Main menu
def main_menu():
    while True:
        print("=====  Welcome to Python Bank =====")
        print("1. Create New Account")
        print("2. Login to Your Account")
        print("3. Exit")

        choice = input("Choose an option: ")
        print()

        if choice == '1':
            create_account()
        elif choice == '2':
            login()
        elif choice == '3':
            print("Thank you for using Python Bank!")
            break
        else:
            print("Invalid choice.\n")

# Run the system
main_menu()
```
